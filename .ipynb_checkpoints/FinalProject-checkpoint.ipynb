# Old player and Young Player Skills Throughout a Career by Michael Condon

## Originally, I was crunching data about ISO, BA, SLG, WOBA, and a new stat I call SpeedRatio, hoping to be able to make a prediction value about how long a player is expected to play based on those carreer stats so far. However, I began to find the correlation between these values and career length was not entirely telling of anything. Instead, I decided to present these findings.

### I used the lahman database from 1950 - 2017. I limited the data to this as to get enough players for the data to mean something, but reduce the number of players with bizzare stats (like Babe Ruth or anyone in the early 1900's). Only years where a batter had at least 445 AB was considered as to reduce the number of unrealistic stats based on little time batting. The CSVs imported were exported from mySQL tables that were made in the FinalProjectDatabaseCreation.sql file. This notebook uses R for statistics.

```r
bestBA = read.csv("bestBAInCareer.csv", sep=";")
bestISO = read.csv("bestISOInCareer.csv", sep=";")
bestSLG = read.csv("bestSLGInCareer.csv", sep=";")
bestWOBA = read.csv("bestWOBAInCareer.csv", sep=";")
bestSpeedRatio = read.csv("bestSpeedRatioInCareer.csv", sep=";")
bestBAInLongCareer = read.csv("bestBAInLongCareer.csv", sep=";")
bestISOInLongCareer = read.csv("bestISOInLongCareer.csv", sep=";")
bestSLGInLongCareer = read.csv("bestSLGInLongCareer.csv", sep=";")
bestWOBAInLongCareer = read.csv("bestWOBAInLongCareer.csv", sep=";")
bestSpeedRatioInLongCareer = read.csv("bestSpeedRatioInLongCareer.csv", sep=";")

playerLongevity = read.csv("playerLongevity.csv", sep=";")

shortCareerPlayers = playerLongevity[(playerLongevity[, "yearsPlayed"] <= 6),]
mediumCareerPlayers = playerLongevity[(playerLongevity$yearsPlayed <= 12 & playerLongevity$yearsPlayed > 6),]
longCareerPlayers = playerLongevity[(playerLongevity[, "yearsPlayed"] > 12),]
```

### One of the most important metrics of measuring a player's Old Player Skills is a player's Isolated Power or ISO. An older player is said to have a better power swing that makes up for his lower speed and batting average. Here we can see a slight trend for the players career lengths and career ISO's.

Mean ISO for Short careers: 0.1274211

Mean ISO for Medium careers: 0.1378429

Mean ISO for Long careers: 0.1669845


Median ISO for Short careers: 0.1251

Median ISO for Medium careers: 0.13365

Median ISO for Long careers: 0.1698

```r
library(ggplot2)

averageIsoForShortCareers = mean(shortCareerPlayers$ISO)
averageIsoForMediumCareers = mean(mediumCareerPlayers$ISO)
averageIsoForLongCareers = mean(longCareerPlayers$ISO)

par(mfrow=c(2,1))
meanIsos <- c(averageIsoForShortCareers, averageIsoForMediumCareers, averageIsoForLongCareers)
barplot(meanIsos, main="Mean ISOs", xlab="Career Lengths", ylab = "ISO", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"), ylim = c(0.0, .20))

averageIsoForShortCareers = median(shortCareerPlayers$ISO)
averageIsoForMediumCareers = median(mediumCareerPlayers$ISO)
averageIsoForLongCareers = median(longCareerPlayers$ISO)

medianIsos <- c(averageIsoForShortCareers, averageIsoForMediumCareers, averageIsoForLongCareers)
barplot(medianIsos, main="Median ISOs", xlab="Career Lengths", ylab = "ISO", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"), ylim = c(0.0, .20))
```

There does not seem to be a large difference between short and medium career lengths with regard to ISO. However, the long career players do seem to have a larger ISO compared to the other two groups. This could be because their above average ISO got them playing for a long time or playing for a long time lead to an above average ISO. 

```r
plot(playerLongevity$yearsPlayed, playerLongevity$ISO, xlab = "Seasons Played", ylab = "Career ISO", main = "Career Length vs Career ISO")
abline(lm(playerLongevity$ISO ~ playerLongevity$yearsPlayed), col="red")
```

Batting Average is considered to be a young player skill because old players tend to hit with more power and young
players tend to get on base more since they are faster. 

```r
library(ggplot2)

averageBAForShortCareers = mean(shortCareerPlayers$BA)
averageBAForMediumCareers = mean(mediumCareerPlayers$BA)
averageBAForLongCareers = mean(longCareerPlayers$BA)

par(mfrow=c(2,1))
meanBAs <- c(averageBAForShortCareers, averageBAForMediumCareers, averageBAForLongCareers)
barplot(meanBAs, main="Mean BAs", xlab="Career Lengths", ylab = "BA", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"), ylim = c(0, .30))

averageBAForShortCareers = median(shortCareerPlayers$BA)
averageBAForMediumCareers = median(mediumCareerPlayers$BA)
averageBAForLongCareers = median(longCareerPlayers$BA)

medianBAs <- c(averageBAForShortCareers, averageBAForMediumCareers, averageBAForLongCareers)
barplot(medianBAs, main="Median BAs", xlab="Career Lengths", ylab = "BA", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"), ylim = c(0, .30))
```

Again, there does not seem to be that strong of a coorelation between career length and batting average. The long 
career players tend to have a higher BA which does not necessarily mean that BA is a old player skill because players
that have gotten to play a large number of seasons tend to be better batters in general compared to ones that had very
short careers. However, calling BA a young skill would not be supported by this data.

```r
plot(playerLongevity$yearsPlayed, playerLongevity$BA, xlab = "Seasons Played", ylab = "Career BA", main = "Career Length vs Career BA")
abline(lm(playerLongevity$BA ~ playerLongevity$yearsPlayed), col="red")
```

SpeedRatio is a stat that I invented. It is said that young players are faster and should therfore be better at getting singles. If a player was to hit with power then he would tend to trade singles for deeper base hits and home runs. SpeedRatio is a stat that is simply the percent of hits that are singles (1B/H). If a player was to have a perfect 1.0 SpeedRatio then he would likely be a weak hitter with very high speed, if a player would have a SpeedRatio of 0.0 than he is either not hitting many balls or all the balls he hits get at least a double indicating power hitting and likely slow speed. 

```r
averagespeedRatioForShortCareers = mean(shortCareerPlayers$speedRatio)
averagespeedRatioForMediumCareers = mean(mediumCareerPlayers$speedRatio)
averagespeedRatioForLongCareers = mean(longCareerPlayers$speedRatio)

par(mfrow=c(2,1))
meanspeedRatios <- c(averagespeedRatioForShortCareers, averagespeedRatioForMediumCareers, averagespeedRatioForLongCareers)
barplot(meanspeedRatios, main="Mean speedRatios", xlab="Career Lengths", ylab = "speedRatio", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"))

averagespeedRatioForShortCareers = median(shortCareerPlayers$speedRatio)
averagespeedRatioForMediumCareers = median(mediumCareerPlayers$speedRatio)
averagespeedRatioForLongCareers = median(longCareerPlayers$speedRatio)

medianspeedRatios <- c(averagespeedRatioForShortCareers, averagespeedRatioForMediumCareers, averagespeedRatioForLongCareers)
barplot(medianspeedRatios, main="Median speedRatios", xlab="Career Lengths", ylab = "speedRatio", 
        names.arg=c("6 Seasons or Less", "7-12 Seasons", "13 or more Seasons"),
        col=c("purple","black", "white"))
```

Yet again, all three groups tended to have pretty similar averages for this stat. It appears there is a slight decrease in speedRatio for players with a long career, but not enough to indicate predictive abilities.

```r
plot(playerLongevity$yearsPlayed, playerLongevity$speedRatio, xlab = "Seasons Played", ylab = "Career speedRatio", main = "Career Length vs Career speedRatio")
abline(lm(playerLongevity$speedRatio ~ playerLongevity$yearsPlayed), col="red")
```

With all this data, I have written an algorithm (below) that takes in ISO, wOBA, BA, SLG, and SpeedRatio and produces all the career lengths of players who were within 10% similarity to these stats. 

```r
readISO <- function()
{ 
  n <- readline(prompt="Enter an ISO: ")
  n <- as.double(n)
  if (is.na(n)){
    n <- readISO()
  }
  return(n)
}
readBA <- function()
{ 
  n <- readline(prompt="Enter an BA: ")
  n <- as.double(n)
  if (is.na(n)){
    n <- readBA()
  }
  return(n)
}
readSLG <- function()
{ 
  n <- readline(prompt="Enter an SLG: ")
  n <- as.double(n)
  if (is.na(n)){
    n <- readSLG()
  }
  return(n)
}
readSpeedRatio <- function()
{ 
  n <- readline(prompt="Enter an SpeedRatio: ")
  n <- as.double(n)
  if (is.na(n)){
    n <- readSpeedRatio()
  }
  return(n)
}
readWOBA <- function()
{ 
  n <- readline(prompt="Enter an WOBA: ")
  n <- as.double(n)
  if (is.na(n)){
    n <- readWOBA()
  }
  return(n)
}

ISO <- readISO()
BA <- readBA()
SpeedRatio <- readSpeedRatio()
SLG <- readSLG()
WOBA <- readWOBA()

ISOUpperBound = ISO * 1.1
ISOLowerBound = ISO * .90

BAUpperBound = BA * 1.1
BALowerBound = BA * .90

SpeedRatioUpperBound = SpeedRatio * 1.1
SpeedRatioLowerBound = SpeedRatio * .90

SLGUpperBound = SLG * 1.1
SLGLowerBound = SLG * .90

WOBAUpperBound = WOBA * 1.1
WOBALowerBound = WOBA * .90

dataLookup = playerLongevity[playerLongevity$ISO >= ISOLowerBound & playerLongevity$ISO <= ISOUpperBound,]
dataLookup = dataLookup[dataLookup$BA >= BALowerBound & dataLookup$BA <= BAUpperBound,]
dataLookup = dataLookup[dataLookup$speedRatio >= SpeedRatioLowerBound & dataLookup$speedRatio <= SpeedRatioUpperBound,]
dataLookup = dataLookup[dataLookup$SLG >= SLGLowerBound & dataLookup$SLG <= SLGUpperBound,]
dataLookup = dataLookup[dataLookup$WOBA >= WOBALowerBound & dataLookup$WOBA <= WOBAUpperBound,]

print(sort(dataLookup$yearsPlayed))

```
